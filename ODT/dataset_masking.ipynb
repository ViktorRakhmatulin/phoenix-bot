```python
import numpy as np
import cv2
import os
from tqdm import tqdm
```

```python
# PATH = 'LINEMOD/test_close/JPEGImages'
PATH = 'test_scene'
```

```python
images_files = [x for x in os.listdir(PATH) if x != 'masks']
images_files.sort(key=lambda x: int(x.split('.')[0]))
len(images_files)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5),(2,2))
def mask_pic(img):
    mask = np.zeros(img.shape[:2], dtype=np.uint8)
    for i, k in enumerate(img):
        for j, _ in enumerate(k):
            if img[i,j,2] > 149 and img[i,j,1] < 115 and img[i,j,0] < 115:
                mask[i,j] = 255
    
    mask = cv2.erode(mask, kernel)
    mask = cv2.dilate(mask, kernel)
    
    contours = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[0]
    
    for c in contours:
        if cv2.contourArea(c) < 400:
            cv2.fillPoly(mask, pts=[c], color=0)
    
    mask = cv2.dilate(mask, kernel, iterations=2)
    mask = cv2.erode(mask, kernel, iterations=3)
    mask = cv2.dilate(mask, kernel)
    
    return mask
```

```python
indx = 0
```

```python
file_name = images_files[indx]
img = cv2.imread(PATH+'/'+file_name)
img = cv2.resize(img, (960, 720), interpolation = cv2.INTER_AREA)

mask = mask_pic(img)
cv2.imshow('orig', img)
cv2.imshow('mask', mask)
cv2.waitKey()
cv2.destroyAllWindows()
cv2.imwrite('test_scene/masks/'+file_name[:-3]+'jpg', mask)
indx +=1
```
